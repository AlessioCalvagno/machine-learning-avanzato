```python
#Import some modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Prima di tutto provo a leggere le immagini di esempio di sklearn e provo ad estrarre hog features da questo... Il resto dopo

```python
from sklearn.datasets import fetch_lfw_people

#size = 64 x 128 as original paper
lfw_people = fetch_lfw_people(resize=1)
```

```python
print(lfw_people.images.shape) #(13233, 62, 47)
img = lfw_people.images[2500]
print(img.min())
print(img.max())
plt.imshow(img, cmap='gray')
plt.show()
```

```python
from skimage.transform import resize
img2 = resize(img,(128,64)) #resize da intendersi come righe e colonne (non come larghezza e altezza)
print(img2.shape)
plt.imshow(img2,cmap='gray')
plt.show()
```

prova estrazione hog features

```python
from skimage.feature import hog
#creating hog features 
#quelli inseriti qui sono i valori usati nel paper originale
fd, hog_image = hog(img2, orientations=9, pixels_per_cell=(4, 4), 
                    cells_per_block=(2, 2), visualize=True, feature_vector=False)
print(fd.shape)
```

```python
plt.subplot(1,2,1)
plt.imshow(img2, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(hog_image, cmap='gray')
plt.show()
```

Se faccio celle più piccole, ottengo una hog_image più dettagliata, nel senso che si riesce a capire meglio il profilo delle forme (1) MA questo succede solo nelle zone dove ho un cambio netto di luminosità (2).

(1) Questo perchè ottengo un numero maggiore di celle e quindi ho una vlautazione più fine nel rilevare le direzioni dei gradienti.

(2) Questo succede perchè faccio normalizzazione più "debole", cioè è fata su aree più piccole e quindi si prendono aree tendenzialmente più uniformi nella luminosità.
